```python
import os
from pydub import AudioSegment
import whisper
from openpyxl import Workbook

def mp3_to_wav(mp3_path):
    audio = AudioSegment.from_mp3(mp3_path)
    wav_path = mp3_path.replace('.mp3', '.wav')
    audio.export(wav_path, format="wav")
    return wav_path

def transcribe_whisper(wav_path):
    model = whisper.load_model("tiny")
    result = model.transcribe(wav_path)
    return result.get("text", "")

def main():
    folder_path = '/content/sample_data/a/a'  # ここに対象フォルダのパスを入れてください
    output_excel_path = 'output.xlsx'

    wb = Workbook()
    ws = wb.active

    for filename in os.listdir(folder_path):
        if filename.endswith('.mp3'):
            full_path = os.path.join(folder_path, filename)
            print(f"Transcribing {filename}...")
            wav_path = mp3_to_wav(full_path)
            transcription = transcribe_whisper(wav_path)
            os.remove(wav_path)
            ws.append([filename, transcription])

    wb.save(output_excel_path)
    print(f"Transcription completed and saved to {output_excel_path}")

if __name__ == "__main__":
    main()
```
